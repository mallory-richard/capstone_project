```python
df = dfs[0]
df.head(20)
```

```python
#If the neighbourhood is not assigned, then I'll use the Borough instead
df['Neighborhood'] = np.where(df['Neighborhood'] == 'Not assigned', df['Borough'], df['Neighborhood'])
```

```python
df.head(20)
```

```python
# Here I am dropping any entries where the Borough was listed as 'Not Assigned'
indexNames = df[ df['Borough'] == 'Not assigned' ].index
# Delete these row indexes from dataFrame
df.drop(indexNames , inplace=True)
df.head(20)
```

```python
#Here I am re-setting the index so that it lists as consecutive numbers
df.reset_index(drop=True, inplace=True)
df.head(20)
```

```python
#Take a look at the csv file while also structuring it as a Pandas dataframe
import csv
latlong = pd.read_csv('latlongdata.csv')
latlong.head(10)
```

```python
#Add the latitude and longitude columns from the csv-derived dataframe to the original dataframe of Toronto postal codes
latitude = latlong['Latitude']
longitude = latlong['Longitude']
df['Latitude']=latitude
df['Longitude']=longitude
df.head(10)
```

## Step 3
Explore the data on a map of Toronto

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinates of Toronto, Ontario, are {}, {}.'.format(latitude, longitude))

```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='red',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
